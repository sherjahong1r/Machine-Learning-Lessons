### 3.3 Transfomer

Yuqorida ma'lumotlarg ishlov berishda biz `sklearn` tarkibidagi tayyor obyektlar va ulardagi `.fit()` yoki `.fit_transform()` funksiyalariga murojaat qildik.
Bu funksiyalar (`imputer`, `OneHotEncoder` va hokazo) **transformer**lar deb ataladi.

Umuman olganda `sklearn` tarkibidagi obyektlar uch turli bo'ladi:
1. **Estimator** - tahmin qiluvchi (baholovchi) funksiyalar. Bunday obyektlar `.fit()` metodiga ega bo'ladi, bu metod berilgan ma'lumotlardan qandaydur qiymatni hisoblaydi (masalan, avvalroq biz `imputer.fit()` yordamida median qiymatni hisobladik)
2. **Transformer** - ham tahmin qilib ham ma'lumotga ishlov beruvchi obyekt. Bunday obyektlar `.fit_transform()` metodiga ega bo'ladi, ya'ni avval qandaydur qiymatnbi hisoblaydi, keyin esa hisoblashni ma'lumotlarga qo'llaydi. Yuoqirdagi `OrdinalEncoder` va `OneHotEncoder` shunday obyektlardan.
3. **Predictor** - bashorat qiluvchi obyektlar. Bunday obyektlar `.predict()` metodiga ega bo'ladi. Ular biz bergan ma'lumotlar asosida bashoratlar qaytaradi. Dars davomida predictor obyektlar bilan ko'p ishlaymiz.

Obyektlarning afzalligi, ular hisob kitob natijalarini ichki attributrlarda saqlab qoladi. Masalan, yuqorida ko'rganimiz `imputer.statistics_`.

Keling endi o'zimiz ham transformer yozishni o'rganamiz.

Esingizda bo'lsa, avvalgi darsimizda biz ma'lumotlarg yangi 2 ustun qo'shdik: `rooms_per_household` va `bedrooms_per_room`.

Keling berilgan ma'lumotlarga shu ikki ustunni avtomat ravishda qo'shadigan transformer yasaymiz.

Buning uchun `sklearn` tarkibida `BaseEstimator` va `TransformerMixin` klasslaridan meros olib yangi klass yasaymiz va bu klassimizga kabi `fit()` va `transform()` metodlarini qo'shamiz:

```python
from sklearn.base import BaseEstimator, TransformerMixin

# bizga kerak ustunlar indexlari
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesadder(BaseEstimator, TransformerMixin):
  def __init__(self, add_bedrooms_per_room=True):
    self.add_bedrooms_per_room = add_bedrooms_per_room
  def fit(self, X, y=None):
    return self # bizning funksiyamiz faqat transformer. estimator emas
  def transform(self, X):
    rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
    population_per_household = X[:, population_ix] / X[:, households_ix]
    if self.add_bedrooms_per_room:  # add_bedrooms_per_room ustuni ixtiyoriy bo'ladi
      bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
      return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
    else:
      return np.c_[X, rooms_per_household, population_per_household]

# from sklearn.base import BaseEstimator, TransformerMixin: Bu qator scikit-learn kutubxonasidan zarur bo'lgan bazaviy
# sinflarni import qiladi. BaseEstimator get_params() va set_params() kabi metodlarni taqdim etadi, TransformerMixin esa
# fit_transform() metodini (u ichkaridan fit() va keyin transform() metodlarini chaqiradi) taqdim etadi.

# rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6: Bu o'zgaruvchilar kirish ma'lumotlaridagi (X) ma'lum
# ustunlarning indeks raqamlarini saqlaydi. Bu transformerga ushbu ustunlarga raqamli indekslari orqali kirish imkonini beradi.

# class CombinedAttributesadder(BaseEstimator, TransformerMixin):: Bu yerda BaseEstimator va TransformerMixin dan meros
# olgan holda maxsus transformer sinfi aniqlanadi. Bu uning scikit-learn ning pipeline xususiyatlari bilan moslashishini ta'minlaydi.

# def __init__(self, add_bedrooms_per_room=True):: Bu konstruktor. U transformerni ishga tushiradi va qo'shimcha xususiyat
# (bedrooms_per_room) qo'shilishi kerakligini belgilash imkonini beradi. Odatiy holda, u True ga o'rnatilgan.

# def fit(self, X, y=None):: Transformer uchun fit metodi odatda o'quv ma'lumotlaridan parametrlarni o'rganadi. Ushbu aniq
# holatda, bu transformer hech qanday parametrni o'rganishi shart emas, shuning uchun u shunchaki self ni qaytaradi.

# def transform(self, X):: Bu yangi xususiyatlar hisoblanadigan va kirish ma'lumotlariga (X) qo'shiladigan asosiy metod. U quyidagi
# hisob-kitoblarni amalga oshiradi:

# rooms_per_household = X[:, rooms_ix] / X[:, households_ix]: Har bir uy xo'jaligidagi xonalar sonini hisoblaydi.
# population_per_household = X[:, population_ix] / X[:, households_ix]: Har bir uy xo'jaligidagi aholi sonini hisoblaydi.
# if self.add_bedrooms_per_room:: Agar add_bedrooms_per_room True bo'lsa (sukut bo'yicha):
# bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]: Har bir xonadagi yotoqxonalar sonini hisoblaydi.
# return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]: U asl ma'lumotlarni
# X ni uchta yangi hisoblangan xususiyat bilan birlashtiradi.
# else:: Agar add_bedrooms_per_room False bo'lsa:
# return np.c_[X, rooms_per_household, population_per_household]: U asl ma'lumotlarni
# X ni faqat ikkita yangi hisoblangan xususiyat bilan birlashtiradi.
```

```python
atter_adder = CombinedAttributesadder()
housing_extra_attribs = atter_adder.transform(housing.values)

# atter_adder = CombinedAttributesadder(): Bu qator CombinedAttributesadder sinfidan atter_adder nomli obyekt yaratadi.
# Bu obyekt ma'lumotlarga yangi xususiyatlarni qo'shish uchun ishlatiladi (masalan, har bir uy xo'jaligidagi xonalar soni kabi).

# housing_extra_attribs = atter_adder.transform(housing.values): Bu qator housing DataFrame'ining qiymatlarini
# (.values orqali NumPy massiviga aylantiriladi) atter_adder obyektiga uzatadi. transform metodi
# yordamida housing.values ma'lumotlariga yangi xususiyatlar qo'shiladi va natija housing_extra_attribs
# o'zgaruvchisiga saqlanadi. Bu yerda housing.values ishlatilishi, transformer matnli ustunlarni kutmasligini bildiradi.
```
